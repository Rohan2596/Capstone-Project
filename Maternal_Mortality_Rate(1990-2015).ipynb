```python
#Plotting Jointplot plot with respect to YEAR=2015  along the 1990,1995,2000,2005
m1=mmr[mmr['1990']<3000]
sns.jointplot(x='1990',y='2015',data=m1,kind='reg')
m2=mmr[mmr['1995']<3000]
sns.jointplot(x='1995',y='2015',data=m2,kind='reg')
m3=mmr[mmr['2000']<3000]
sns.jointplot(x='2000',y='2015',data=m3,kind='reg')
m4=mmr[mmr['2005']<3000]
sns.jointplot(x='2005',y='2015',data=m4,kind='reg')
```

# Machine Learning
Building an model to predict the clusters labels
Dividing the countries into 7 cluster using kmeans clustering algorithms

```python
X=mmr[['1990','1995','2000','2005','2010','2015']]
```

```python
#Creating KMeans Clustering 
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=7,random_state=0).fit(X)
mmr['kmeans1']=kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
mmr['kmeans1'].value_counts().sort_index().plot.bar(figsize=(14,14),fontsize=7)
plt.xlabel('kmeans1')
plt.ylabel('Countries/Regions')
print(mmr['kmeans1'].value_counts().sort_index())
plt.style.use('ggplot')
```

# Density based clustering 

```python
from sklearn.cluster import DBSCAN
Dbscan=DBSCAN(eps=3,min_samples=2).fit(X)
mmr['Dbscan1']=Dbscan.labels_
```

```python
mmr
```

```python
mmr['Dbscan1'].value_counts().sort_index().plot.bar(figsize=(14,14),fontsize=7)
plt.xlabel('kmeans1')
plt.ylabel('Countries/Regions')
plt.style.use('ggplot')
mmr['Dbscan1'].value_counts()
```
